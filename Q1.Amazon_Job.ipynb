ii)Correalation Matrix 

```python
import seaborn as sns





# Convert 'Work Life Balance' and other columns to numeric (if needed)
numeric_columns = ['Overall Rating', 'Work Life Balance', 'Career Opportunities']

# Handle missing values (e.g., fill NaN with 0 or mean)
data[numeric_columns] = data[numeric_columns].fillna(0)

# Calculate correlation matrix
corr_matrix = data[numeric_columns].corr()

print("Correlation Matrix:")
print(corr_matrix)

# Optional: Visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

4. Visualise at least two interesting inferences, or information that you may want to draw attention to, from
the chosen dataset using your favourite visualisation tool.

```python


# Plot the distribution of overall ratings
plt.figure(figsize=(10, 6))
data['Overall ratting'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.xlabel('Overall Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Overall Ratings')
plt.xticks(rotation=0)
plt.show()
```

```python
#  proportion of each overall rating
rating_counts = data['Overall ratting'].value_counts()


plt.figure(figsize=(10, 8))#pie chart
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Proportion of Overall Ratings')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```

```python
# Scatter plot of Work/Life Balance vs. Overall Rating
plt.figure(figsize=(10, 6))
plt.scatter(data['Work/Life Balance'], data['Overall ratting'], alpha=0.6, color='purple')
plt.xlabel('Work/Life Balance')
plt.ylabel('Overall Rating')
plt.title('Work/Life Balance vs. Overall Rating')
plt.grid(True)
plt.show()
```
